```python
import pandas as pd
import matplotlib.pyplot as plt

weather_data = {
    "Day": list(range(1, 31)),
    "Temperature": [22, 23, 25, 24, 26, 28, 30, 29, 27, 26,
                    25, 24, 23, 22, 24, 25, 27, 29, 31, 30,
                    28, 27, 26, 25, 24, 23, 22, 24, 26, 27]
}
weather_df = pd.DataFrame(weather_data)
weather_df.to_csv("daily_temperature.csv", index=False)

plt.figure(figsize=(10,5))
plt.plot(weather_df["Day"], weather_df["Temperature"], marker='o', color='orange', linestyle='-')
plt.title("Daily Temperature Changes (Weather Dataset)")
plt.xlabel("Day")
plt.ylabel("Temperature (°C)")
plt.grid(True)
plt.show()

hottest_day = weather_df.loc[weather_df["Temperature"].idxmax()]
coldest_day = weather_df.loc[weather_df["Temperature"].idxmin()]

print("🌞 Hottest Day:", hottest_day["Day"], "with", hottest_day["Temperature"], "°C")
print("❄️ Coldest Day:", coldest_day["Day"], "with", coldest_day["Temperature"], "°C")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

sales_data = {
    "Quarter": ["Q1", "Q2", "Q3", "Q4"],
    "Sales": [15000, 12000, 18000, 25000]
}
sales_df = pd.DataFrame(sales_data)
sales_df.to_csv("sales_data.csv", index=False)

plt.figure(figsize=(8,5))
plt.plot(sales_df["Quarter"], sales_df["Sales"], marker='o', color='blue', linestyle='-')
plt.title("Quarterly Sales Performance (Retail Sales Dataset)")
plt.xlabel("Quarter")
plt.ylabel("Sales")
plt.grid(True)

peak_idx = sales_df["Sales"].idxmax()
plt.scatter(sales_df["Quarter"][peak_idx], sales_df["Sales"][peak_idx], color='red', s=100, label="Peak Sales")
plt.legend()
plt.show()

print("📈 Peak Sales in", sales_df["Quarter"][peak_idx], "with", sales_df["Sales"][peak_idx])
```
